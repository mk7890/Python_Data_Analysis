```python
# Replace specific values
credit_customers['checking_status'] = credit_customers['checking_status'].replace({'no checking': 'new_value'})
```

```python
credit_customers['employment'].value_counts()
```

```python
employment_mapping = {
    '<1' : 'Low Income',
    '1<=X<4': 'Medium income',
    '4<=X<7' : 'High income',
    '>=7': 'Very High Income'
}
credit_customers['employment'] = credit_customers['employment'].replace(employment_mapping)
credit_customers
```

```python
credit_customers['savings_status'].value_counts()
```

```python
savings_status_mapping= {
    'no known savings' : 'No Savings',
    '<100' : 'Low Savings',
    '100<=X<500' : 'Medium Savings',
    '>=500' : 'High Savings',
    '>=1000' : 'Very High Savings'
}
credit_customers['savings_status'] = credit_customers['savings_status'].replace(savings_status_mapping)
credit_customers
```

```python
checking_status_mapping = {
    'no checking' : 'Low Balance',
    '<0' : 'Negative Balance',
    '0<=X<200' : 'Medium Checking',
    '>=200' : 'High Balance'
}
credit_customers['checking_status'] = credit_customers['checking_status'].replace(checking_status_mapping)
credit_customers
```

```python
credit_customers['credit_history'].value_counts()
```

```python
credit_customers['credit_history'] = credit_customers['credit_history'].replace({
    'critical/other existing credit' : 'Very poor credit history',
    'existing paid' : 'Poor Credit History',
    'delayed previously' : 'Medium Credit History',
    'all paid' : 'Good Credit History',
    'no credits/all paid' : 'Excellent Credit History'
})
credit_customers
```
