```python
# GeoportailFrance.orthos
attribution = '<a target="_blank" href="https://www.geoportail.gouv.fr/">Geoportail France</a>'


tiles = 'https://data.geopf.fr/wmts?REQUEST=GetTile&SERVICE=WMTS&VERSION=1.0.0&STYLE={style}&TILEMATRIXSET=PM&FORMAT={format}&LAYER=ORTHOIMAGERY.ORTHOPHOTOS&TILEMATRIX={z}&TILEROW={y}&TILECOL={x}'

myMap = folium.Map(location=[kenya_lat, kenya_lon], zoom_start=zoom_start)

folium.TileLayer(
    tiles = tiles,
    attr = attribution,
    min_zoom=2,
    max_zoom=19,
    format='image/jpeg',
    style = 'normal'
).add_to(myMap)
display(myMap)
```

```python
import folium
# Geoportail France Orthophotos Tile Layer URL and Attribution
tiles = 'https://data.geopf.fr/wmts?REQUEST=GetTile&SERVICE=WMTS&VERSION=1.0.0&STYLE={style}&TILEMATRIXSET=PM&FORMAT={format}&LAYER=ORTHOIMAGERY.ORTHOPHOTOS&TILEMATRIX={z}&TILEROW={y}&TILECOL={x}'
attr = '<a target="_blank" href="https://www.geoportail.gouv.fr/">Geoportail France</a>'
# Create a map centered on a specific location (e.g., Paris, France)
m = folium.Map(location=(48.8566, 2.3522), zoom_start=6)
# Add the Geoportail France orthophotos tile layer
folium.TileLayer(
    tiles=tiles,
    attr=attr,
    min_zoom=2,
    max_zoom=19,
    format='image/jpeg',
    style='normal'
).add_to(m)
# Display the map
m
```

# Adding Markers

```python
kenya_lat = 1.2921
kenya_lon = 36.8219
attr = (
    '&copy; <a href="https://www.openstreetmap.org/copyright">OpenStreetMap</a> contributors'
)

tiles = 'https://tile.openstreetmap.org/{z}/{x}/{y}.png'

kenya_map = folium.Map(location=[kenya_lat, kenya_lon], tiles=tiles, attr=attr, zoom_start=zoom_start)

folium.Marker(
    location=[kenya_lat, 36.0000],
    tooltip="Click me!",
    popup="Mt. Hood Meadows",
    icon=folium.Icon(icon="cloud"),
).add_to(kenya_map)

folium.Marker(
    location=[kenya_lat, kenya_lon],
    tooltip="Click me!",
    popup="Timberline Lodge",
    icon=folium.Icon(color="green"),
).add_to(kenya_map)

display(kenya_map)
kenya_map.save("kenya_map2.html")
```
