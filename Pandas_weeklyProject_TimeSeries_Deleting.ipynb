- 09_Time_Series/Investor_Flow_of_Funds_US/Exercise


```python
# Step 1. Import the necessary libraries
import numpy as np
import pandas as pd
```

```python
# Step 2. Import the dataset from this address. :https://raw.githubusercontent.com/datasets/investor-flow-of-funds-us/master/data/weekly.csv
# Step 3. Assign it to a variable called
funds_url='https://raw.githubusercontent.com/datasets/investor-flow-of-funds-us/master/data/weekly.csv'
funds=pd.read_csv(funds_url)
funds.head()
funds.to_csv('invester_flow_of_funds_us.csv') # save funds csv file
```

```python
# Step 4. What is the frequency of the dataset?
funds.info()
```

```python
# Step 5. Set the column Date as the index.
funds=funds.set_index('Date')
funds
```

```python
# Step 6. What is the type of the index?
funds.index.dtype
```

```python
# Step 7. Set the index to a DatetimeIndex type
funds.index=pd.to_datetime(funds.index)
funds.index.dtype
```

```python
# Step 8. Change the frequency to monthly, sum the values and assign it to monthly.
monthly=funds.resample('M').sum()
monthly.columns
monthly
```

```python
monthly.columns
```

```python
# Step 9. You will notice that it filled the dataFrame with months that don't have any data with NaN. Let's drop these rows.
monthly.drop(monthly[(monthly == 0).any(axis=1)].index, inplace=True)
```

```python
monthly
```

```python
# Step 10. Good, now we have the monthly data. Now change the frequency to year.
yearly=monthly.resample('YE').sum()
yearly
```

- 10_Deleting/Wine  dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import dataset from address and assign to variable
wine_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'
wine = pd.read_csv(wine_url)
wine
```

```python
wine.columns
```

```python
# Delete the first, fourth, seventh, nineth, eleventh, thirteenth and fourteenth columns
# Assuming your DataFrame is named df
columns_to_delete = ['1', '2.43', '2.8', '.28', '5.64', '3.92', '1065']
wine = wine.drop(columns=columns_to_delete)

# Print the updated DataFrame
print(wine)
```

```python
# Assign the columns as below:
'''
The attributes are (donated by Riccardo Leardi, riclea '@' anchem.unige.it):

alcohol
malic_acid
alcalinity_of_ash
magnesium
flavanoids
proanthocyanins
hue
'''
wine.columns=['alcohol','malic_acid','alcalinity_of_ash','magnesium','flavanoids','proanthocyanins','hue']
wine
```

```python
# Set the values of the first 3 rows from alcohol as NaN
wine.loc[:2,'alcohol']=np.nan
wine
```

```python
# Now set the value of the rows 3 and 4 of magnesium as NaN
wine.loc[2:3,'magnesium']=np.nan
wine
```

```python
# Fill the value of NaN with the number 10 in alcohol and 100 in magnesium
wine['alcohol'].fillna(10,inplace=True)
wine['magnesium'].fillna(100,inplace=True)
wine
```

```python
# Count the number of missing values
wine.isnull().sum()
```

```python
# Create an array of 10 random numbers up until 10
random_numbers=np.random.randint(0,10,10)
random_numbers
```

```python
# Use random numbers you generated as an index and assign NaN value to each of cell
wine.loc[random_numbers]=np.nan
wine
```

```python
# How many missing values do we have?
wine.isnull().value_counts().sum()
```

```python
wine.isnull().sum()
```

```python
# Delete the rows that contain missing values
wine.dropna(inplace=True)
wine
```

```python
# Print only the non-null values in alcohol
wine[wine['alcohol'].notnull()]
```

```python
# Reset the index, so it starts with 0 again
wine.reset_index(drop=True,inplace=True)
wine
```
