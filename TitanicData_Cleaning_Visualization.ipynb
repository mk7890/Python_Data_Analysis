```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

titanic = pd.read_csv('/content/Titanic-Dataset.csv')
titanic
```

```python
titanic.info()
```

```python
titanic.describe()
```

```python
titanic.columns
```

```python
titanic.isnull().sum()
```

```python
percentage = (titanic.isnull().sum()/len(titanic))*100
percentage
```

```python
titanic.columns
```

```python
titanic=pd.DataFrame(titanic, columns=['Name', 'Sex', 'Age', 'Survived', 'Pclass', 'SibSp', 'Parch', 'Ticket',
       'Fare', 'Cabin', 'Embarked'])
titanic
```

```python
titanic['Cabin'].value_counts()
```

```python
titanic = titanic.drop(['Cabin'],axis=1)
titanic
```

```python
titanic['Embarked'].value_counts()
```

```python
titanic['Embarked']=titanic['Embarked'].replace({
    'S':'Southampton',
    'C':'Cherbourg',
    'Q':'Queenstown'
})
titanic['Embarked']
```

```python
titanic.isnull().sum()
```

```python
titanic['Age'].fillna(titanic['Age'].mean(), inplace=True)
titanic['Age'] = titanic['Age'].astype(int)
titanic['Age']
```

```python
titanic['Embarked'].fillna(titanic['Embarked'].mode()[0],inplace=True)
titanic['Embarked']
```

```python
titanic.isnull().sum()
```

```python
titanic.columns
```

```python
plt.figure(figsize=(5, 7))
sns.boxplot(y='Age', data=titanic)
plt.title('Passenger Age Boxplot')

```

```python
plt.figure(figsize=(10, 5))
plt.hist(titanic['Age'], bins=10, color='skyblue', edgecolor='black')
plt.title('Titanic Data')
plt.xlabel('Age')
plt.ylabel('Survived')
plt
```

```python
sns.kdeplot(titanic['Age'], fill=True)
plt.title('Titanic Passanger Age Data')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.violinplot(x=titanic['Sex'], y=titanic['Age'])
plt.title('Titanic Passenger Sex vs Age')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.violinplot(x=titanic['Sex'], y=titanic['Age'], hue=titanic['Survived'])
plt.title('Titanic Passenger Sex vs Age')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```

#Sex vs Survived

```python
sns.kdeplot(titanic['Age'][titanic.Survived==True], fill=True, label='Survived')
sns.kdeplot(titanic['Age'][titanic.Survived==False], fill=True, label='Dead')
plt.legend()
plt.title('Titanic Survived Data')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

# Violin plot for survived, age, and class

```python
sns.violinplot(x=titanic['Pclass'], y=titanic['Age'], hue=titanic['Survived'], split=True, data=titanic)
plt.title('Violin Plot for Survived, Age, and Class')
plt.xlabel('Pclass')
plt.ylabel('Age')
plt.show()
```

```python
sns.kdeplot(titanic['Age'][titanic.Sex=='male'], fill=True, label='Male')
sns.kdeplot(titanic['Age'][titanic.Sex=='female'], fill=True, label='Female')
plt.legend()
plt.title('Titanic Survived Data')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```
