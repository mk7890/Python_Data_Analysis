```python
# import necessary packages
import numpy as np
import pandas as pd
```

```python
# Step 2. Import the dataset.
# Step 3. Assign it to a variable called crime.
crime_rawData_url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv'
crime = pd.read_csv(crime_rawData_url)
crime.head()
```

```python
# Step 4. What is the type of the columns?
crime.info()
```

```python
# Have you noticed that the type of Year is int64. But pandas has a different type to work with Time Series. Let's see it now.
# Step 5. Convert the type of the column Year to datetime64
crime['Year'] = pd.to_datetime(crime['Year'], format='%Y')
crime.info()
```

```python
# Step 6. Set the Year column as the index of the dataframe
crime = crime.set_index('Year', drop=True)
crime.head()
```

```python
# Step 7. Delete the Total column
del crime['Total']
crime.head()
```

```python
# Step 8. Group the year by decades and sum the values. Pay attention to the Population column number, summing this column is a mistake

# Convert the index to datetime format and extract only the year
crime.index = pd.to_datetime(crime.index).year

# Create a new column for decades based on the index
crime['Decade'] = (crime.index // 10) * 10

# Group by the 'Decade' column and sum all columns except 'Population'
summed_by_decade = crime.groupby('Decade').sum(numeric_only=True)

# For the 'Population' column, take the maximum value within each decade
summed_by_decade['Population'] = crime.groupby('Decade')['Population'].max()

# Display the result
print(summed_by_decade)


```

```python
# Step 9. What is the most dangerous decade to live in the US?


# Convert the index to datetime format and extract only the year
crime.index = pd.to_datetime(crime.index).year

# Create a new column for decades based on the index
crime['Decade'] = (crime.index // 10) * 10

# Group by the 'Decade' column and sum all columns except 'Population'
summed_by_decade = crime.groupby('Decade').sum(numeric_only=True)

# For the 'Population' column, take the maximum value within each decade
summed_by_decade['Population'] = crime.groupby('Decade')['Population'].max()

# Calculate the total crimes per decade (excluding 'Population')
summed_by_decade['Total_Crimes'] = summed_by_decade.drop(columns='Population').sum(axis=1)

# Find the decade with the highest total crimes
most_dangerous_decade = summed_by_decade['Total_Crimes'].idxmax()
most_dangerous_value = summed_by_decade['Total_Crimes'].max()

# Display the most dangerous decade and the crime count
print(f"Most dangerous decade: {most_dangerous_decade} with {most_dangerous_value} total crimes")

# Display the summarized DataFrame
print(summed_by_decade)
```

```python
# Step 1. Import the necessary libraries
import numpy as np
import pandas as pd
```

```python
# Step 2. Create the 3 DataFrames based on the following raw data
raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}

raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
```

```python
# Step 3. Assign each to a variable called data1, data2, data3
data1 = pd.DataFrame(raw_data_1)
print('data1\n',data1)
data2 = pd.DataFrame(raw_data_2)
print('data2\n', data2)
data3 = pd.DataFrame(raw_data_3)
print('data3\n',data3)
```

```python
# Step 4. Join the two dataframes along rows and assign all_data (data1 & data2)
all_data = pd.concat([data1, data2], axis=0)
all_data
```

```python
# Step 5. Join the two dataframes along columns and assing to all_data_col
all_data_col = pd.concat([data1, data2], axis=1)
all_data_col
```

```python
# Step 6. Print data3
data3
```

```python
# Step 7. Merge all_data and data3 along the subject_id value
all_data_data3 = pd.merge(all_data, data3, on='subject_id')
all_data_data3
```

```python
# Step 8. Merge only the data that has the same 'subject_id' on both data1 and data2
data1_data2 = pd.merge(data1, data2, on='subject_id')
data1_data2
```

```python
# Step 9. Merge all values in data1 and data2, with matching records from both sides where available.
data1_data2_all = pd.merge(data1, data2, on='subject_id', how='outer')
data1_data2_all
```
