```python
# Step 10. For the most-ordered item, how many items were ordered?
print(f"{chipo['item_name'].value_counts().idxmax()} ordered items : {chipo['item_name'].value_counts().max()}")
```

```python
# Step 11. What was the most ordered item in the choice_description column?
print(f"Most ordered item in the choice_description column : {chipo['choice_description'].value_counts().idxmax()}")
```

```python
# Step 12. How many items were orderd in total?
print(f"items were orderd in total : {chipo['quantity'].sum()}")
```

```python
# Step 13. Turn the item price into a float
#chipo['item_price'] = chipo['item_price'].str.replace('$', '').astype(float)
chipo['item_price'] = chipo['item_price'].replace('[\$,]', '', regex=True).astype(float)
chipo['item_price']
```

```python
# Step 13.a. Check the item price type
chipo['item_price'].dtype
```

```python
# Step 13.b. Create a lambda function and change the type of item price

chipo['item_price'] = chipo['item_price'].apply(lambda x: float(str(x).replace('$', '').replace(',', '')))

# Display the updated DataFrame
chipo.head()
```

```python
# Step 13.c. Check the item price type
chipo['item_price'].dtype
```

```python
# Step 14. How much was the revenue for the period in the dataset?
revenue = (chipo['quantity'] * chipo['item_price']).sum()
print('Revenue was: $' + str(np.round(revenue,2)))
```

```python
# Step 15. How many orders were made in the period?
orders = chipo['order_id'].value_counts().count()
print(f"orders made in the period : {orders}")
```

```python
# Step 16. What is the average revenue amount per order?
average_revenue_per_order = revenue / orders
print(f"average revenue amount per order : ${average_revenue_per_order}")
```

```python
# Step 17. How many different items are sold?
print(f"different items are sold : {chipo['item_name'].nunique()}")
```

```python
baby_names['Name'].nunique()
```

```python
# What is the name with most occurrences?
names['Name'].value_counts().idxmax()
```

```python
# How many different names have the least occurrences?

# Count the occurrences of each name
name_counts = baby_names['Name'].value_counts()

# Find the minimum count
min_count = name_counts.min()

# Find how many names have this minimum count
num_least_occurrences = (name_counts == min_count).sum()
print("Number of names with the least occurrences:", num_least_occurrences)
```

```python
# What is the median name occurrence?
names['Name'].value_counts().median()
```

```python
# Count the occurrences of each name
name_counts = baby_names['Name'].value_counts()

# Calculate the median occurrence
median_occurrence = name_counts.median()

# Print the result
print("Median name occurrence:", median_occurrence)
```

```python
# What is the standard deviation of names?
names['Name'].value_counts().std()
```

```python
# Get a summary with the mean, min, max, std and quartiles.

# Count the occurrences of each name
name_counts = baby_names['Name'].value_counts()

# Get the summary statistics
summary_stats = name_counts.describe()

# Print the result
print(summary_stats)
```

- 06_Stats/Wind_Stats dataset

```python
# Step 1. Import the necessary libraries
import numpy as np
import pandas as pd
```

```python
# Step 2. Import the dataset from this address : https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data
# Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index.
wind_url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data'
data = pd.read_csv(wind_url, sep='\s+', parse_dates=[[0, 1, 2]])
data
```

```python
data.columns
```

```python
data.info()
```

```python
# Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

# Function to replace year starting with 20 by 19
def replace_year(x):
    if x.year >= 2000 and x.year < 2100:
        return x.replace(year=x.year - 100)  # Subtract 100 from the year
    return x

# Apply the replace_year function to the 'Yr_Mo_Dy' column
data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(replace_year)

# Display the updated DataFrame
data
```

```python
# Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].
data.set_index('Yr_Mo_Dy', inplace=True)
data
```

```python
# Step 6. Compute how many values are missing for each location over the entire record.
data.isnull().sum()
```

```python
# Step 7. Compute how many non-missing values there are in total.

total_non_missing_values = data.count().sum()
print(total_non_missing_values)
```

```python
# Step 8. Calculate the mean windspeeds of the windspeeds over all the locations and all the times.
data.mean().mean()
```

```python
# Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days
loc_stats = pd.DataFrame({
    'Min': data.min(),
    'Max': data.max(),
    'Mean': data.mean(),
    'Std': data.std()
})
loc_stats
```

```python
# Step 10. Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day.
day_stats = pd.DataFrame({
    'Min': data.min(axis=1),
    'Max': data.max(axis=1),
    'Mean': data.mean(axis=1),
    'Std': data.std(axis=1)
})
day_stats
```

```python
# Step 11. Find the average windspeed in January for each location
average_january_winds = data[data.index.month == 1].mean()
average_january_winds
```

```python
# Step 12. Downsample the record to a yearly frequency for each location.
yearly_data = data.resample('Y').mean()
yearly_data
```

```python
# Step 13. Downsample the record to a monthly frequency for each location.
monthly_data = data.resample('M').mean()
monthly_data
```

```python
# Step 14. Downsample the record to a weekly frequency for each location.
weekly_data = data.resample('W').mean()
weekly_data
```

```python
# Step 15. Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week (assume that the first week starts on January 2 1961) for the first 52 weeks

weekly_stats = pd.DataFrame({
    'Min': weekly_data.min(axis=1),
    'Max': weekly_data.max(axis=1),
    'Mean': weekly_data.mean(axis=1),
    'Std': weekly_data.std(axis=1)
})
day_stats
```
