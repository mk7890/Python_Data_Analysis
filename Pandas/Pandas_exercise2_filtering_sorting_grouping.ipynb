# 02_Filtering_&_Sorting/Chipotle

```python
# chipotle
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(url, sep = '\t')
chipo
```

```python
chipo.info()
```

```python
chipo.describe()
```

```python
chipo.columns
```

```python
# How many products cost more than $10.00?

# clean the item_price column and transform it in a float
chipo['item_price'] = chipo['item_price'].replace('[\$,]', '', regex=True).astype(float)  # Convert prices to float
products_above_10 = chipo[chipo['item_price'] > 10]['item_name'].nunique()

print(f"Number of products costing more than $10.00: {products_above_10}")
```

```python
# What is the price of each item?

# Select only the 'item_name' and 'item_price' columns
price_df = chipo[['item_name', 'item_price']]

# Print the resulting DataFrame
print(price_df)
```

```python
# Sort by the name of the item


# Select only the 'item_name' and 'item_price' columns
price_df = chipo[['item_name', 'item_price']]

# Sort the DataFrame by the 'item_name' column
sorted_price_df = price_df.sort_values(by='item_name')

# Print the resulting sorted DataFrame
print(sorted_price_df)
```

```python
# What was the quantity of the most expensive item ordered?


most_expensive_item = chipo.loc[chipo['item_price'].idxmax()]

# Extract the quantity of that item
quantity_of_most_expensive_item = most_expensive_item['quantity']

print(f"Quantity of the most expensive item ordered: {quantity_of_most_expensive_item}")
```

```python
# How many times was a Veggie Salad Bowl ordered?

# Filter the rows where 'item_name' is "Veggie Salad Bowl"
veggie_salad_orders = chipo[chipo['item_name'] == 'Veggie Salad Bowl']

# Count the number of orders (rows)
num_veggie_salad_orders = len(veggie_salad_orders)

print(f"Veggie Salad Bowl was ordered {num_veggie_salad_orders} times.")
```

```python
# How many times did someone order more than one Canned Soda?
# Filter orders with 'Canned Soda' and quantity greater than 1
canned_soda_orders = chipo[(chipo['item_name'] == 'Canned Soda') & (chipo['quantity'] > 1)]

# Count the number of such orders
num_orders = canned_soda_orders['order_id'].nunique()

print(f"Number of times someone ordered more than one Canned Soda: {num_orders}")
```

# 3. Grouping

- Alcohol Consmption

```python
alcohol_url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv'
drinks = pd.read_csv(alcohol_url)
drinks
```

```python
drinks.columns
```

```python
# Which continent drinks more beer on average?

# Group the data by continent and calculate the average beer servings
average_beer_servings = drinks.groupby('continent')['beer_servings'].mean()

# Find the continent with the highest average beer servings
continent_with_most_beer = average_beer_servings.idxmax()
highest_average_beer = average_beer_servings.max()

print(f"The continent that drinks the most beer on average is {continent_with_most_beer} with an average of {highest_average_beer} beer servings.")
```

```python
# For each continent print the statistics for wine consumption.

# Group by continent and get statistics for wine_servings
wine_statistics = drinks.groupby('continent')['wine_servings'].describe()

# Print the statistics for each continent
print(wine_statistics)
```

```python
# Print the mean alcohol consumption per continent for every column


# Group by continent and calculate the mean for every relevant column
mean_alcohol_consumption = drinks.groupby('continent')[['beer_servings', 'spirit_servings', 'wine_servings', 'total_litres_of_pure_alcohol']].mean()

# Print the mean consumption per continent for each column
print(mean_alcohol_consumption)
```

```python
# Print the median alcohol consumption per continent for every column

# Group by continent and calculate the median for every relevant column
median_alcohol_consumption = drinks.groupby('continent')[['beer_servings', 'spirit_servings', 'wine_servings', 'total_litres_of_pure_alcohol']].median()

# Print the median consumption per continent for each column
print(median_alcohol_consumption)
```

```python
# Print the mean, min and max values for spirit consumption. This time output a DataFrame


# Group by continent and calculate the mean, min, and max for spirit_servings
spirit_consumption_stats = drinks.groupby('continent')['spirit_servings'].agg(['mean', 'min', 'max'])

# Print the resulting DataFrame
print(spirit_consumption_stats)
```

- 03_Grouping/Occupation/Exercise

```python
# import necessary libraries
import numpy as np
import pandas as pd
```

```python
# Step 2. Import the dataset
# Step 3. Assign it to a variable called users.
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
users = pd.read_csv(url, sep='|')
users
```

```python
# Step 4. Discover what is the mean age per occupation
mean_age_per_occupation = users.groupby('occupation')['age'].mean()
mean_age_per_occupation
```

```python
# Step 5. Discover the Male ratio per occupation and sort it from the most to the least
male_ratio_per_occupation = users[users['gender'] == 'M'].groupby('occupation')['gender'].count() / users.groupby('occupation')['gender'].count() * 100
male_ratio_sorted = male_ratio_per_occupation.sort_values(ascending=False)
male_ratio_sorted
```

```python
# Step 6. For each occupation, calculate the minimum and maximum ages
min_max_age_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])
min_max_age_per_occupation
```

```python
# Step 7. For each combination of occupation and gender, calculate the mean age
mean_age_per_occupation_gender = users.groupby(['occupation', 'gender'])['age'].mean()
mean_age_per_occupation_gender
```

```python
# Step 8. For each occupation present the percentage of women and men
gender_percentage_per_occupation = users.groupby('occupation')['gender'].value_counts(normalize=True) * 100
gender_percentage_per_occupation
```
