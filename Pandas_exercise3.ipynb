# Student Alcohol Consumption

```python
url='https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/Students_Alcohol_Consumption/student-mat.csv'
df=pd.read_csv(url)
df.head()
```

```python
df.columns
```

```python
# For the purpose of this exercise slice the dataframe from 'school' until the 'guardian' column


# Slice the DataFrame from 'school' to 'guardian'
sliced_df = df.loc[:, 'school':'guardian']

# Print the resulting DataFrame
sliced_df
```

```python
# Create a lambda function that will capitalize strings.

# Apply lambda function to capitalize strings in the sliced DataFrame
capitalized_sliced_df = sliced_df.applymap(lambda x: x.capitalize() if isinstance(x, str) else x)

# Print the resulting DataFrame
capitalized_sliced_df
```

```python
# Capitalize both Mjob and Fjob

# Capitalize the values in the 'Mjob' and 'Fjob' columns
capitalized_sliced_df['Mjob'] = capitalized_sliced_df['Mjob'].apply(lambda x: x.capitalize() if isinstance(x, str) else x)
capitalized_sliced_df['Fjob'] = capitalized_sliced_df['Fjob'].apply(lambda x: x.capitalize() if isinstance(x, str) else x)

# Print the updated DataFrame
capitalized_sliced_df[['Mjob', 'Fjob']]
```

```python
# Print the last elements of the data set.
capitalized_sliced_df.tail()
```

```python
# Did you notice the original dataframe is still lowercase? Why is that? Fix it and capitalize Mjob and Fjob.

# Capitalize all string values in the DataFrame
sliced_df = sliced_df.applymap(lambda x: x.capitalize() if isinstance(x, str) else x)

# Now capitalize Mjob and Fjob explicitly (optional, just to be sure)
sliced_df['Mjob'] = sliced_df['Mjob'].apply(lambda x: x.capitalize() if isinstance(x, str) else x)
sliced_df['Fjob'] = sliced_df['Fjob'].apply(lambda x: x.capitalize() if isinstance(x, str) else x)

# Print the updated DataFrame
sliced_df[['Mjob', 'Fjob']]
```

```python
# Create a function called majority that returns a boolean value to a new column
# called legal_drinker (Consider majority as older than 17 years old)

# Define the majority function
def majority(age):
    return age > 17

# Apply the function to the 'age' column and create a new 'legal_drinker' column
sliced_df['legal_drinker'] = sliced_df['age'].apply(majority)

# Print the updated DataFrame with the new 'legal_drinker' column
sliced_df[['age', 'legal_drinker']]
```

```python
# Multiply every number of the dataset by 10.

# Multiply every numerical value in the DataFrame by 10
sliced_df = sliced_df.applymap(lambda x: x * 10 if isinstance(x, (int, float)) else x)

# Print the updated DataFrame
sliced_df
```

# 4. Merge

- Auto MPG Cars dataset

```python
import numpy as np
import pandas as pd
```

```python
# Import the first dataset cars1 and cars2.

cars1_url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv'
cars2_url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv'

cars1 = pd.read_csv(cars1_url)
cars2 = pd.read_csv(cars2_url)
```

```python
cars1
```

```python
cars2
```

```python
# Oops, it seems our first dataset has some unnamed blank columns, fix cars1

cars1 = cars1.loc[:, 'mpg':'car']
cars1
```

```python
cars1.columns
```

```python
cars2.columns
```

```python
# What is the number of observations in each dataset?

# Get the number of observations (rows) in the DataFrame
num_observations_cars1 = cars1.shape[0]
num_observations_cars2 = cars2.shape[0]

# Print the result
print("Number of observations cars1 :", num_observations_cars1)
print("Number of observations cars2 :", num_observations_cars2)
```

```python
# Join cars1 and cars2 into a single DataFrame called cars


# Concatenate the two DataFrames along rows (axis=0)
cars = pd.concat([cars1, cars2], axis=0, ignore_index=True)

# Print the resulting DataFrame
cars
```

```python
# Oops, there is a column missing, called owners. Create a random number Series from 15,000 to 73,000.
# Add the column owners to cars

# Create a random number Series for the 'owners' column
cars['owners'] = np.random.randint(15000, 73001, size=len(cars))

# Print the resulting DataFrame
cars
```
